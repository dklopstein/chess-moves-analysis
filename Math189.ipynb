Now that we have some preliminary understanding of our data, we can move onto further analysis and discovery regarding the questions we would like to answer. Firstly, lets take a look at the relationship between the color that a player is playing and the win result/their rating. To do so we will perform a t-test for indendence 

**Null Hypothesis($H_0$) - There is no significant difference in ratings between white and black players.**

**Alternate Hypothesis($H_1$)- There is a significant difference in ratings between white and black players.**

````python
white_ratings = df[df['winner'] == 'white']['white_rating']
black_ratings = df[df['winner'] == 'black']['black_rating']

t_stat, p_value = ttest_ind(white_ratings, black_ratings)
print(f'T-Statistic: {t_stat}\nP-Value: {p_value}')
````

Since the p-value is not less than 0.05, we fail to reject the null hypothesis. In other words, there is no significant difference in ratings between white and black players.


Next, lets take a look whether or not the type of opening move has a significant impact on victory of the game. To do this we can perform a Chi-square Test of Independence to determine whether there is a significant association between these two categorical variables(opening move and win result)


**Null Hypothesis($H_0$)- The type of opening moves does not significantly affect victory status.**

**Alternate Hypothesis($H_1$)- The type of opening moves does significantly affect victory status.**

````python
contingency_table = pd.crosstab(df['opening_eco'], df['victory_status'])

# Perform the chi-square test
chi2_stat, p_value, _, _ = chi2_contingency(contingency_table)

print(f'Chi-Square Statistic: {chi2_stat}\nP-Value: {p_value}')
````

Since our p-value is really small, we can safely reject the null hypothesis. In other words, the type of opening moves does significantly affect victory status, and we intend to dive further into this idea.

````python
#identifies the top 10 most frequently occurring opening moves
top_10_openings = df['opening_eco'].value_counts().head(10).index

#filter the dataset to only have top 10 opening moves
df_top_10 = df[df['opening_eco'].isin(top_10_openings)]

#calculates win percentage for each opening move
win_percentage_by_opening = (
    df_top_10.groupby('opening_eco')['victory_status']
    .value_counts(normalize=True)
    .unstack(fill_value=0)
    .loc[:, 'mate':'resign']
)

#Visualize win percentage for each opening move
sns.heatmap(win_percentage_by_opening, annot=True, cmap="RdYlGn_r", fmt=".2%")
plt.title('Win Percentage by Opening Move')
plt.xlabel('Victory Status')
plt.ylabel('Opening Move')
plt.show()
````

From here we can perform the Kruskal-Wallis non-parametric test that can be used to determine whether there are statistically significant differences between the distributions of three or more independent groups. In this case the ten groups would be the ten most popular opening moves.


**Null Hypothesis($H_0$)- There is no significant difference in win percentages among the top 10 opening moves.**

**Alternate Hypothesis($H_1$)- There is a significant difference in win percentages among the top 10 opening moves.**

````python
#get win percentages for each opening move
opening_win_percentages = [
    df_top_10[df_top_10['opening_eco'] == opening]['victory_status'].value_counts(normalize=True)['mate']
    for opening in top_10_openings
]

#perform Kruskal-Wallis test
h_stat, p_value_kruskal = kruskal(*opening_win_percentages)

#display the results
print(f'Kruskal-Wallis H-Statistic: {h_stat}\nP-Value: {p_value_kruskal}')
#display mate percentages 
print(opening_win_percentages)
````

Since the p-value is greater than our chosen siginificacne of 0.05, we fail to reject the null hypothesis. In other words, there is no significant difference in win percentages among the top 10 opening moves.
